```python
# 07_backtest_simulation.ipynb

import pandas as pd
import numpy as np
import os
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

# --- CONFIG ---
DATA_DIR = "../data"
MODELS_DIR = "../models"
VAULT_SELL_TEST = os.path.join(DATA_DIR, "Vault_Test_Sell.parquet")
MODEL_SELL_PATH = os.path.join(MODELS_DIR, "RF_Sell_HailMary.joblib")

# Threshold from your analysis
CONFIDENCE_THRESHOLD = 0.70
RISK_PER_TRADE = 1.0   # 1% risk per trade (Loss = -1R)
REWARD_PER_TRADE = 1.5 # 1.5% profit per trade (Win = +1.5R)

# --- SIMULATION ENGINE ---
def run_simulation():
    print(f"üîÑ Loading Vault Data from {VAULT_SELL_TEST}...")
    if not os.path.exists(VAULT_SELL_TEST):
        print("‚ùå Vault file missing.")
        return

    # Load Data & Model
    df_test = pd.read_parquet(VAULT_SELL_TEST).dropna()
    model = joblib.load(MODEL_SELL_PATH)
    
    # 1. Get Probabilities
    # We need the specific "Confidence Score" for every candle
    # The model outputs [Prob_Class_0, Prob_Class_1]
    features = ['EMA_50_Slope', 'Dist_from_200', 'RSI', 'ATR', 'Rel_Vol', 'hour']
    probs = model.predict_proba(df_test[features])[:, 1]
    
    # 2. Filter Trades
    # Create a DataFrame of ONLY the trades we would have taken
    df_test['confidence'] = probs
    trades = df_test[df_test['confidence'] >= CONFIDENCE_THRESHOLD].copy()
    
    if len(trades) == 0:
        print("üíÄ No trades taken at this threshold.")
        return

    # 3. Calculate Outcomes
    # If Label=1, we won (+1.5R). If Label=0, we lost (-1.0R).
    trades['pnl_r'] = np.where(trades['LABEL_SELL'] == 1, REWARD_PER_TRADE, -RISK_PER_TRADE)
    
    # 4. Compute Equity Curve
    trades['equity_curve'] = trades['pnl_r'].cumsum()
    
    # 5. Calculate Drawdown
    trades['peak'] = trades['equity_curve'].cummax()
    trades['drawdown'] = trades['equity_curve'] - trades['peak']
    max_drawdown = trades['drawdown'].min()
    
    # 6. Stats
    total_trades = len(trades)
    wins = len(trades[trades['pnl_r'] > 0])
    losses = len(trades[trades['pnl_r'] < 0])
    win_rate = wins / total_trades
    net_profit = trades['equity_curve'].iloc[-1]
    
    # --- REPORT CARD ---
    print(f"\nüìä BACKTEST RESULTS (Threshold {CONFIDENCE_THRESHOLD})")
    print(f"{'='*40}")
    print(f"üóìÔ∏è  Duration:      ~1 Year (Vault Data)")
    print(f"üî´  Total Trades:  {total_trades}")
    print(f"‚úÖ  Win Rate:      {win_rate:.2%}")
    print(f"üí∞  Net Profit:    {net_profit:+.2f}R")
    print(f"üìâ  Max Drawdown:  {max_drawdown:.2f}R")
    print(f"üßä  Expectancy:    {net_profit / total_trades:+.2f}R per trade")
    
    # --- VISUALIZATION ---
    plt.figure(figsize=(12, 8))
    
    # Plot 1: Equity Curve
    plt.subplot(2, 1, 1)
    plt.plot(range(len(trades)), trades['equity_curve'], marker='o', linestyle='-', color='green')
    plt.axhline(0, color='black', linestyle='--', alpha=0.5)
    plt.title(f"Equity Curve: Sell Sniper (Threshold {CONFIDENCE_THRESHOLD})")
    plt.ylabel("Profit (R-Multiples)")
    plt.grid(True, alpha=0.3)
    
    # Plot 2: Drawdown
    plt.subplot(2, 1, 2)
    plt.fill_between(range(len(trades)), trades['drawdown'], 0, color='red', alpha=0.3)
    plt.title("Drawdown Analysis")
    plt.ylabel("Drawdown (R)")
    plt.xlabel("Trade Number")
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

    # Show the first 10 trades to verify logic
    print("\nüîç Recent Trades Log:")
    print(trades[['confidence', 'LABEL_SELL', 'pnl_r']].tail(10))

# Run it
run_simulation()
```
