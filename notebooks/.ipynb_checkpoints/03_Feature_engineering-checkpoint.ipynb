### Imports & Load Labeled Data

```python
import pandas as pd
import numpy as np
# import pandas_ta as ta
import matplotlib.pyplot as plt

# Load your LABELED dataset (ensure you saved the 0.8 version)
df = pd.read_parquet("../data/EURUSD_D1_Labeled.parquet")

print(f"Loaded {len(df)} rows with columns: {df.columns.tolist()}")
```

### Add Technical Indicators (The "Evidence")

```python
# --- REPLACEMENT CODE: NO PANDAS_TA REQUIRED ---

# the mighty rsi- relative strength index
def calculate_rsi(series, period=14):
    # todayys close - yesterday's close--  did we go up or down
    delta = series.diff(1)

    # how strong have the bulls been lately
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    
    # hows strong have the bears been in the past 2 weeks in average
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()

    # noamrlising the relative strength betweeen 0 and  100
    rs = gain / loss

    # as  rs is high, the bulls are exhausted, else the bears are
    return 100 - (100 / (1 + rs))

def add_features(df):
    df = df.copy()
    
    # --- A. MOMENTUM (RSI) ---
    # Manual RSI Calculation (Standard Formula) -- the usual indicators of velocity
    df['rsi'] = calculate_rsi(df['close'], period=14)
    df['rsi_slope'] = df['rsi'].diff(1)
    df['rsi_accel'] = df['rsi_slope'].diff(1)
    
    # --- B. TREND (SMA) ---
    # Simple Moving Averages
    df['sma_50'] = df['close'].rolling(window=50).mean()
    df['sma_200'] = df['close'].rolling(window=200).mean()
    
    # Distance from MA
    df['dist_sma50'] = (df['close'] - df['sma_50']) / df['sma_50']
    df['dist_sma200'] = (df['close'] - df['sma_200']) / df['sma_200']
    
    # --- C. VOLATILITY (Bollinger Bands) ---
    # Manual BB Calculation
    # this is essentially for the last 20 trading days approx
    window = 20
    # the normal distribution
    std_dev = 2.0
    # the general trendove the last 20 days
    rolling_mean = df['close'].rolling(window=window).mean()
    # the volatility meter, ie how jumpy the price has been over 20 days
    rolling_std = df['close'].rolling(window=window).std()

    # the actual bollinger bands
    upper_band = rolling_mean + (rolling_std * std_dev)
    lower_band = rolling_mean - (rolling_std * std_dev)
    
    # Band Width: (Upper - Lower) / Middle
    df['bb_width'] = (upper_band - lower_band) / rolling_mean
    # Band Position: (Price - Lower) / (Upper - Lower)
    df['bb_position'] = (df['close'] - lower_band) / (upper_band - lower_band)
    
    # ATR Ratio (Relative Volatility)
    # Note: We already calculated 'atr' in the previous notebook, so it should exist.
    # If not, we can approximate it with High-Low.
    if 'atr' in df.columns:
        df['atr_rel'] = df['atr'] / df['atr'].rolling(30).mean()
    else:
        # Fallback approximation
        tr = df['high'] - df['low']
        df['atr_rel'] = tr.rolling(14).mean() / tr.rolling(30).mean()
    
    # --- D. TIME / SEASONALITY ---
    df['day_of_week'] = df.index.dayofweek
    df['month'] = df.index.month
    
    # --- E. PRICE ACTION (Lagged Returns) ---
    # this is the percentage change in the price
    for lag in [1, 2, 3, 5]:
        df[f'return_t-{lag}'] = df['close'].pct_change(lag)
    
    return df

print("üõ†Ô∏è Engineering Features (Manual Mode)...")
df_features = add_features(df)

# Drop NaN values (Must do this because indicators need history to warm up)
df_features.dropna(inplace=True)

print(f"‚úÖ Features added. Final dataset shape: {df_features.shape}")
display(df_features.tail(3))
```

### Apply the "Veto" Filters

```python
def apply_veto_filters(df, max_spread_points):
    """
    Forces labels to 0 (Wait) if specific 'Do Not Trade' conditions are met.
    """
    df_filtered = df.copy()
    
    # --- 0. COUNT BEFORE ---
    # We capture the state of the strategy BEFORE the vetoes
    orig_buys = len(df_filtered[df_filtered['label'] == 1])
    orig_sells = len(df_filtered[df_filtered['label'] == -1])
    
    # --- 1. SPREAD VETO ---
    if 'spread' in df_filtered.columns:
        # this is the mask for removing the spreads
        high_spread_mask = df_filtered['spread'] > max_spread_points
        # locate all the rows where the mask is true and change the label to wait
        df_filtered.loc[high_spread_mask, 'label'] = 0
        print(f"üö´ Vetoed {high_spread_mask.sum()} days due to High Spread (> {max_spread_points} pts).")
    
    # --- 2. NFP VETO ---
    # if it is friday
    is_friday = df_filtered.index.dayofweek == 4
    # if it is also the first friday of the month
    is_first_week = df_filtered.index.day <= 7
    # create the mask
    nfp_mask = is_friday & is_first_week

    # apply the nfp mask
    df_filtered.loc[nfp_mask, 'label'] = 0
    print(f"üö´ Vetoed {nfp_mask.sum()} days due to Likely NFP (First Friday).")

    # --- 3. DECEMBER VETO ---
    december_mask = df_filtered.index.month == 12
    df_filtered.loc[december_mask, 'label'] = 0
    print(f"üö´ Vetoed {december_mask.sum()} days due to December (Low Liquidity).")
    
    # --- 4. COUNT AFTER ---
    new_buys = len(df_filtered[df_filtered['label'] == 1])
    new_sells = len(df_filtered[df_filtered['label'] == -1])
    
    # Stats
    removed_buys = orig_buys - new_buys
    removed_sells = orig_sells - new_sells
    
    print("\n--- SAFETY VETO REPORT ---")
    print(f"üìâ Buys Removed:  {removed_buys} (From {orig_buys} to {new_buys})")
    print(f"üìâ Sells Removed: {removed_sells} (From {orig_sells} to {new_sells})")
    print(f"Total Trades Sacrificed: {removed_buys + removed_sells}")
    
    return df_filtered
```

### The "Golden Rule" Check (No Peeking!)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Apply the Veto Filters (High Spread, NFP, December)
# We use the updated function that counts both Buys and Sells
print("üõ°Ô∏è Applying Safety Vetoes...")
df_vetoed = apply_veto_filters(df_features, max_spread_points=25)

# 2. Check Class Balance (Did we kill too many trades?)
print("\n--- FINAL CLASS BALANCE ---")
print(df_vetoed['label'].value_counts(normalize=True).sort_index())

# 3. Correlation Sanity Check
# We want to make sure 'label' isn't 99% correlated with any feature (Leakage)
# We also want to see if our features still have SOME signal (Non-zero correlation)
print("\nüîç Checking Correlations on Vetoed Data...")
correlations = df_vetoed.corr(numeric_only=True)['label'].sort_values()

print("\nTop Correlations with Target:")
print(correlations)
```

### #  Visual Heatmap (Optional - for top 10 features)

```python
top_features = correlations.abs().nlargest(11).index # Top 10 + label
plt.figure(figsize=(10, 8))
sns.heatmap(df_vetoed[top_features].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (Vetoed Data)")
plt.show()
```

### 5. SAVE THE FINAL DATASET

```python
# This is the "Gold Master" file for the XGBoost Model
output_path = "../data/EURUSD_D1_Ready_for_XGBoost_Vetoed.parquet"
df_vetoed.to_parquet(output_path)

print(f"\n‚úÖ SUCCESS: Final dataset saved to {output_path}")
print("üöÄ Ready for Phase 3: Model Training!")
```
