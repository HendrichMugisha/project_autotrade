### Feature Selection & Initial Classification

```python
# Cell 1: Imports
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# Configuration
# If you are continuing directly from the previous cell, use df_feat.
# If loading from disk, uncomment the line below:
df_feat = pd.read_parquet("../data/engineered_features_16h.parquet")

print(f"Data Shape: {df_feat.shape}")
```

### Data Preparation (X/y Split & Label Mapping)

```python
# Cell 2: Prepare X and y

# 1. Define Target
y = df_feat['label'].copy()

# 2. Map Labels for XGBoost (It requires 0, 1, 2)
# Current: -1 (Sell), 0 (Range), 1 (Buy)
# Mapped:   0 (Sell), 1 (Range), 2 (Buy)
label_mapping = {-1: 0, 0: 1, 1: 2}
y_mapped = y.map(label_mapping)

# 3. Define Features (X)
# Drop the target and raw price columns (keep only derived features)
# NOTE: We keep 'close' only if used for normalization later, but usually we drop raw prices
drop_cols = ['label', 'open', 'high', 'low', 'close', 'tick_volume', 'volume', 'spread', 'atr']
# intersection allows us to drop only what exists to avoid errors
cols_to_drop = [c for c in drop_cols if c in df_feat.columns]

X = df_feat.drop(columns=cols_to_drop)

print(f"Features (X) Shape: {X.shape}")
print(f"Target (y) Shape: {y_mapped.shape}")
print(f"Class Balance:\n{y_mapped.value_counts(normalize=True)}")
```

### Time-Series Split (Chronological)

```python
# Cell 3: Chronological Split

# We'll use the first 80% for training, last 20% for validation
split_point = int(len(X) * 0.80)

X_train = X.iloc[:split_point]
X_test = X.iloc[split_point:]

y_train = y_mapped.iloc[:split_point]
y_test = y_mapped.iloc[split_point:]

print(f"Training Samples: {len(X_train)} (Start: {X_train.index[0]} - End: {X_train.index[-1]})")
print(f"Testing Samples:  {len(X_test)}  (Start: {X_test.index[0]} - End: {X_test.index[-1]})")
```

### MODEL TRAINIGN

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
from sklearn.metrics import classification_report

# --------------------------------------------------------------------------------
# 1. Define the Parameter Grid
# --------------------------------------------------------------------------------
# We test "balanced" (punish mistakes) vs None (standard)
# We test Depth to find the balance between "Memorizing Noise" and "Learning Trends"
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, 20, None],  # None = fully grown trees (risky but powerful)
    'min_samples_split': [2, 10, 50], # Higher number = less noise, more stability
    'class_weight': ['balanced', None], # Crucial: Let the grid decide if punishment helps
    'criterion': ['gini', 'entropy']
}

# --------------------------------------------------------------------------------
# 2. Setup Time-Series Cross Validation
# --------------------------------------------------------------------------------
# Standard CV shuffles data. In trading, that is "cheating" (using next week's price).
# TimeSeriesSplit ensures we only train on past data and test on future data.
tscv = TimeSeriesSplit(n_splits=3)

# --------------------------------------------------------------------------------
# 3. Initialize the Grid Search
# --------------------------------------------------------------------------------
rf = RandomForestClassifier(random_state=42)

# We optimize for 'precision_weighted' because you want safer trades, 
# not just high accuracy (which you can get by just predicting "Range" forever)
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=tscv,
    scoring='precision_weighted', # optimize for QUALITY of predictions
    # n_jobs=-1, # Use all CPU cores
    verbose=1
)

print("Starting Grid Search... (This may take a minute)")
grid_search.fit(X_train, y_train)

# --------------------------------------------------------------------------------
# 4. Results & Best Parameters
# --------------------------------------------------------------------------------
print(f"\nBest Parameters Found: {grid_search.best_params_}")
print(f"Best CV Score (Precision Weighted): {grid_search.best_score_:.4f}")

# --------------------------------------------------------------------------------
# 5. Final Evaluation on Test Set
# --------------------------------------------------------------------------------
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

print("\n--- FINAL TEST SET RESULTS ---")
print(classification_report(y_test, y_pred, target_names=['Sell', 'Range', 'Buy']))

# Optional: Feature Importance (To see if your features are actually useful)
import pandas as pd
import matplotlib.pyplot as plt

# If X_train is a DataFrame, use column names; else just use indices
feat_names = X_train.columns if hasattr(X_train, 'columns') else [f"Feat_{i}" for i in range(X_train.shape[1])]

importances = pd.Series(best_model.feature_importances_, index=feat_names)
print("\n--- Top 10 Most Important Features ---")
print(importances.sort_values(ascending=False).head(10))
```
