### Load the Dream Dataset

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from scipy.stats import randint
import joblib

# 1. SETUP PATHS
INPUT_FILENAME = "dream_features_12h.parquet" 
INPUT_PATH = os.path.join('../data', INPUT_FILENAME)

# 2. LOAD DATA
if os.path.exists(INPUT_PATH):
    print(f"Loading dataset from {INPUT_PATH}...")
    df = pd.read_parquet(INPUT_PATH)
    print(f"Dataset Shape: {df.shape}")
else:
    raise FileNotFoundError(f"CRITICAL: File not found at {INPUT_PATH}. Please check the filename.")
```

### Prepare Training Data (X & y)

```python
# 1. SEPARATE FEATURES AND TARGET
target_col = 'label'
X = df.drop(columns=[target_col])
y = df[target_col]

# 2. MAP LABELS (Standardizing to 0, 1, 2)
# -1 (Sell) -> 0
#  0 (Hold) -> 1
#  1 (Buy)  -> 2
label_mapping = {-1: 0, 0: 1, 1: 2}
y_mapped = y.map(label_mapping)

# 3. ALIGN DATA (Safety check for dropped rows)
common_index = X.index.intersection(y_mapped.index)
X_aligned = X.loc[common_index]
y_aligned = y_mapped.loc[common_index]

# 4. CREATE THE "VAULT" (Holdout Set)
# We split strictly by time. The last 20% is locked away.
split_point = int(len(X_aligned) * 0.80)

X_train = X_aligned.iloc[:split_point]
y_train = y_aligned.iloc[:split_point]

X_test_vault = X_aligned.iloc[split_point:]
y_test_vault = y_aligned.iloc[split_point:]

print(f"--- SPLIT COMPLETE ---")
print(f"Training Set : {len(X_train)} rows (Used for Grid Search)")
print(f"Vault Set    : {len(X_test_vault)} rows (Hidden for Final Test)")
print(f"\nTraining Class Balance:\n{y_train.value_counts(normalize=True)}")
```

### Configure the Optimization (The "Thorough" Search)

```python
# 1. VALIDATION STRATEGY (Internal CV on Training Data)
# 5 Splits with a 50-step gap to prevent overlap
tscv = TimeSeriesSplit(n_splits=5, gap=50)

# 2. SEARCH SPACE (Aggressive / High-Signal Config)
param_dist = {
    # Trees: 250-1000 for stability
    'n_estimators': randint(250, 1000),
    
    # Depth: 8-20 (Allowing deeper trees to find specific setups)
    'max_depth': randint(8, 20),
    
    # Leaf Size: 15-100 (Slightly lower min to catch more signals)
    'min_samples_leaf': randint(15, 100),
    
    # Randomness
    'max_features': ['sqrt', 'log2', None],

    'bootstrap': [True, False],
    
    # Class Weight: Critical test. 
    # 'None' = let majority rule. 'balanced' = help the minority Hold.
    'class_weight': ['balanced', 'balanced_subsample', None],
    
    # Split Quality
    'criterion': ['gini', 'entropy']
}

print("Configuration Ready: Optimization will run on Training Set only.")
```

### Execute Random Search

```python
# 1. INITIALIZE MODEL
rf = RandomForestClassifier(random_state=42)

# 2. INITIALIZE SEARCH
search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=150,                   # 50 iterations for speed (bump to 100 for precision)
    scoring='precision_weighted',          # Focus on balanced performance across Sell/Hold/Buy
    cv=tscv,
    verbose=1,
    random_state=42,
    # n_jobs=1                     # Single core to prevent Windows crash
)

# 3. RUN TRAINING
print(f"Starting optimization on {len(X_train)} rows...")
search.fit(X_train, y_train)

print("\n--- OPTIMIZATION COMPLETE ---")
print(f"Best CV Score (Macro F1): {search.best_score_:.4f}")
print("Best Parameters:", search.best_params_)
```

### Final Evaluation & Feature Importance

### The "Honest Test" (Evaluation)

```python
# 1. GET THE WINNER
best_model = search.best_estimator_

# 2. PREDICT ON VAULT DATA
print("Running prediction on Vault (Holdout) data...")
y_pred_vault = best_model.predict(X_test_vault)

# 3. CLASSIFICATION REPORT
target_names = ['Sell', 'Hold', 'Buy'] # 0, 1, 2
print("\n--- ðŸ”’ HONEST TEST RESULTS ---")
print(classification_report(y_test_vault, y_pred_vault, target_names=target_names))

# 4. CONFUSION MATRIX
cm = confusion_matrix(y_test_vault, y_pred_vault)
print("\nConfusion Matrix:")
print(cm)
```

### Load Data 

```python
import joblib
import os
import pandas as pd

def load_model_bundle(filepath):
    if not os.path.exists(filepath):
        raise FileNotFoundError(f"Bundle not found at: {filepath}")
    
    print(f"--- ðŸ”“ Opening Vault: {filepath} ---")
    bundle = joblib.load(filepath)
    
    # Extract components
    model = bundle['model']
    features_expected = bundle['feature_names']
    
    # Print the metadata you saved
    print(f" > Horizon:   {bundle.get('horizon', 'Unknown')}")
    print(f" > Strategy:  {bundle.get('strategy', 'Unknown')}")
    print(f" > Note:      {bundle.get('description', 'No notes')}")
    print(f" > Rec Thresh:{bundle.get('threshold_recommendation', 'N/A')}")
    
    return model, features_expected, bundle
```

### Feature Alignment (Crucial Safety Step

```python
def align_features(X_test, expected_features):
    # 1. Check for missing columns
    missing = set(expected_features) - set(X_test.columns)
    if missing:
        raise ValueError(f"ðŸš¨ ALERT: The Vault Test Set is missing features required by the model: {missing}")
    
    # 2. Reorder columns to match training exactly
    X_aligned = X_test[expected_features]
    print(f"âœ… Features aligned. Shape: {X_aligned.shape}")
    return X_aligned
```

### The Evaluation Suite (Confidence, Importance, Buy/Sell)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

def run_diagnostics(model, X_test, y_test, label_mapping):
    
    # --- A. PREDICTIONS ---
    probs = model.predict_proba(X_test)
    preds = model.predict(X_test)
    
    # --- B. GRANULAR REPORT ---
    reverse_map = {v: k for k, v in label_mapping.items()} # 0->Sell, etc
    names = [f"{reverse_map[0]} (Sell)", f"{reverse_map[1]} (Hold)", f"{reverse_map[2]} (Buy)"]
    
    print("\n--- 1. Granular Buy vs Sell Analysis ---")
    cm = confusion_matrix(y_test, preds)
    
    # Precision Check
    sell_prec = cm[0,0] / cm[:,0].sum() if cm[:,0].sum() > 0 else 0
    buy_prec  = cm[2,2] / cm[:,2].sum() if cm[:,2].sum() > 0 else 0
    
    print(f"TRUST METRIC (Precision):")
    print(f" > When model says SELL: {sell_prec:.2%} Correct")
    print(f" > When model says BUY : {buy_prec:.2%} Correct")
    
    # Confusion Matrix Plot
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=names, yticklabels=names)
    plt.title('Confusion Matrix')
    plt.show()

    # --- C. CONFIDENCE DISTRIBUTION ---
    print("\n--- 2. Confidence Check ---")
    conf_df = pd.DataFrame({
        'Predicted': preds,
        'Max_Conf': probs.max(axis=1)
    })
    plt.figure(figsize=(8, 3))
    sns.histplot(data=conf_df, x='Max_Conf', hue='Predicted', bins=30, palette='viridis', element="step")
    plt.title("How confident is the model in its choices?")
    plt.show()

    # --- D. FEATURE IMPORTANCE ---
    print("\n--- 3. What drove the decisions? ---")
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
        feats = X_test.columns
        indices = np.argsort(importances)[::-1][:15] # Top 15
        
        plt.figure(figsize=(8, 5))
        sns.barplot(x=importances[indices], y=feats[indices], palette='magma')
        plt.title('Top 15 Feature Importances')
        plt.show()
```

### Execution Block

```python
# 1. DEFINE FILENAME (Copy/Paste the exact filename you just saved)
target_file = 'honest_model_12h_horizon_20251231_1418.joblib' # <--- UPDATE THIS

# 2. LOAD BUNDLE
model_loaded, expected_feats, bundle_info = load_model_bundle(target_file)

# 3. ALIGN VAULT DATA (Critical!)
# Ensures X_test_vault matches the exact column order the model was trained on
X_vault_ready = align_features(X_test_vault, expected_feats)

# 4. RUN EVALUATION
run_diagnostics(model_loaded, X_vault_ready, y_test_vault, label_mapping)
```

### The Threshold Analysis Function

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score
def analyze_thresholds(model, X_test, y_test, label_mapping):
    print("--- ðŸŽšï¸ Running Sensitivity Analysis (Precision vs Volume) ---")
    
    probs = model.predict_proba(X_test)
    max_probs = probs.max(axis=1)
    preds = model.predict(X_test)
    true_labels = y_test.values if hasattr(y_test, 'values') else y_test
    
    # Define thresholds
    thresholds = np.arange(0.35, 0.56, 0.025)
    results = []
    
    for thresh in thresholds:
        mask = max_probs >= thresh
        
        # If no trades, record 0s but keep Threshold as FLOAT
        if mask.sum() == 0:
            results.append({
                'Threshold': float(thresh),  # <--- FIX: Keep as float
                'Total_Actives': 0, 
                'Sell_Count': 0, 'Buy_Count': 0,
                'Sell_Prec': 0.0, 'Buy_Prec': 0.0
            })
            continue
            
        y_filt_true = true_labels[mask]
        y_filt_pred = preds[mask]
        
        prec = precision_score(y_filt_true, y_filt_pred, average=None, labels=[0,1,2], zero_division=0)
        
        n_sells = (y_filt_pred == 0).sum()
        n_buys  = (y_filt_pred == 2).sum()
        
        results.append({
            'Threshold': float(thresh),  # <--- FIX: Keep as float
            'Total_Actives': n_sells + n_buys,
            'Sell_Count': n_sells,
            'Buy_Count': n_buys,
            'Sell_Prec': prec[0],
            'Buy_Prec':  prec[2]
        })
        
    df_res = pd.DataFrame(results)
    
    # Print formatted table
    print("\n--- ðŸ“Š Threshold Impact Table ---")
    print(df_res.to_string(index=False, float_format="%.3f"))
    
    # Plot
    fig, ax1 = plt.subplots(figsize=(10, 6))
    ax1.plot(df_res['Threshold'], df_res['Buy_Prec'], 'g-o', label='Buy Precision')
    ax1.plot(df_res['Threshold'], df_res['Sell_Prec'], 'r-o', label='Sell Precision')
    ax1.set_xlabel('Confidence Threshold')
    ax1.set_ylabel('Precision (Win Rate)')
    ax1.legend(loc='upper left')
    ax1.grid(True, alpha=0.3)
    
    ax2 = ax1.twinx()
    ax2.bar(df_res['Threshold'], df_res['Total_Actives'], alpha=0.1, color='gray', width=0.02)
    ax2.set_ylabel('Trade Count')
    
    plt.title("Sensitivity Analysis: Precision vs Volume")
    plt.show()
    return df_res
```

```python
df_sensitivity = analyze_thresholds(model_loaded, X_vault_ready, y_test_vault, label_mapping)
```
