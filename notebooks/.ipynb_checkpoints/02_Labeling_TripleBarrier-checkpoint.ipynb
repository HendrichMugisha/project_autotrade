### Imports & Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm

# 1. Load the Raw H1 Data
df = pd.read_parquet("../data/EURUSD_H1_Raw.parquet")


# this is a standard candles range
high_low = df['high'] - df['low']
high_prev_close = abs(df['high'] - df['close'].shift(1))
low_prev_close = abs(df['low'] - df['close'].shift(1))

# Create a DataFrame of the 3 components and take the max
tr_df = pd.concat([high_low, high_prev_close, low_prev_close], axis=1)
# pick the max of all the three
df['tr'] = tr_df.max(axis=1)

# we need to get an average
# using exponential weighted moving average to give more weright to recent data
df['atr'] = df['tr'].ewm(alpha=1/14, min_periods=14, adjust=False).mean()

# Drop the first 14 rows (NaNs)
df.dropna(subset=['atr'], inplace=True)

print(f"âœ… Loaded {len(df)} candles.")
print(f"   Current H1 ATR (Volatility): {df['atr'].iloc[-1]:.5f} points")
display(df.head(3))
```

### __The Triple Barrier Logic (The Core)_ 

```python
def apply_h1_triple_barrier(df, horizon_hours=12, atr_multiplier=2.0):
    """
    Labels each hour for Intraday Swing Trading.
    
    Logic:
    - Horizon: Look ahead 'horizon_hours' (e.g., 12 hours).
    - Barrier: 'atr_multiplier' * Current Volatility.
    
    Returns:
    1  (Buy):  Price hit Upper Barrier first.
    -1 (Sell): Price hit Lower Barrier first.
    0  (Wait): Price hit neither (Choppy/Range) or ran out of time.
    """
    labels = []
    
    # Convert columns to numpy arrays for speed
    close_arr = df['close'].values
    high_arr = df['high'].values
    low_arr = df['low'].values
    atr_arr = df['atr'].values
    
    # Loop through the dataset
    for i in tqdm(range(len(df) - horizon_hours), desc="Labeling H1 Swings"):

        # hypothethical entry
        entry_price = close_arr[i]
        # how wild is the market rn?
        volatility = atr_arr[i]
        
        # 1. Set Dynamic Barriers
        barrier = volatility * atr_multiplier
        # tp for a buy
        upper_limit = entry_price + barrier
        # tp for a sell
        lower_limit = entry_price - barrier
        
        # 2. Look Ahead (Slice the next 12 candles)
        # this is the list of future highs
        future_highs = high_arr[i+1 : i+horizon_hours+1]
        # this is a lis of future lows
        future_lows = low_arr[i+1 : i+horizon_hours+1]
        
        # 3. Check for Touches
        # did we hit any? if so which ones -- returns theindixes
        hit_upper_indices = np.where(future_highs >= upper_limit)[0]
        # the very first one that hit
        first_upper_hit = hit_upper_indices[0] if len(hit_upper_indices) > 0 else 999
        
        hit_lower_indices = np.where(future_lows <= lower_limit)[0]
        first_lower_hit = hit_lower_indices[0] if len(hit_lower_indices) > 0 else 999
        
        # 4. Determine Winner based on time
        if first_upper_hit == 999 and first_lower_hit == 999:
            labels.append(0) # Time Stop (Range)

            # which one happened first
        elif first_upper_hit < first_lower_hit:
            labels.append(1) # Bullish Win
        elif first_lower_hit < first_upper_hit:
            labels.append(-1) # Bearish Win
        else:
            labels.append(0) # Tie / Noise
            
    # Pad the end
    labels.extend([np.nan] * horizon_hours)
    
    return labels

print("âœ… Labeling Function Defined.")
```

### Execution (Settings)

```python
# SETTINGS
HORIZON = 6     # 12 Hours hold time
MULTIPLIER = 1.5   # Target = 2.0x ATR

print(f"ðŸš€ Running Triple Barrier (Horizon={HORIZON}h, Target={MULTIPLIER}x ATR)...")

df['label'] = apply_h1_triple_barrier(df, horizon_hours=HORIZON, atr_multiplier=MULTIPLIER)

# Clean up
df_clean = df.dropna(subset=['label']).copy()
df_clean['label'] = df_clean['label'].astype(int)

# --- REPORT CARD ---
print("\n--- Class Balance (H1 Data) ---")
print(df_clean['label'].value_counts(normalize=True).sort_index())
print(f"\nTotal labeled candles: {len(df_clean)}")
```

### Visualization (NEW)

```python
# --- VISUALIZATION ---
# Plot last 500 candles (approx 3 weeks on H1) to see the detail
candles_to_plot = 100

# Grab subset
subset = df.dropna(subset=['label']).tail(candles_to_plot)

plt.figure(figsize=(15, 6))
plt.plot(subset.index, subset['close'], alpha=0.5, color='gray', label='Price')

# Plot Buys (Green Up Triangles)
buys = subset[subset['label'] == 1]
plt.scatter(buys.index, buys['close'], color='green', marker='^', s=40, label='Buy (1)', zorder=5)

# Plot Sells (Red Down Triangles)
sells = subset[subset['label'] == -1]
plt.scatter(sells.index, sells['close'], color='red', marker='v', s=40, label='Sell (-1)', zorder=5)

# Plot Range (Blue Dots)
ranges = subset[subset['label'] == 0]
plt.scatter(ranges.index, ranges['close'], color='blue', s=10, alpha=0.3, label='Wait (0)')

plt.title(f"Triple Barrier Labels horizon: {HORIZON}, multiplier: {MULTIPLIER}, (Last {candles_to_plot} Hours)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
