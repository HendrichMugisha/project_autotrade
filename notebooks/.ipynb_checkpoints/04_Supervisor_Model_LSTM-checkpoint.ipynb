### Functional API Model Architecture

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, LSTM, Bidirectional, Dropout, BatchNormalization, Conv1D, GlobalAveragePooling1D, MultiHeadAttention, LayerNormalization, Add, Activation
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# --- 1. SOTA COMPONENTS ---

def build_sota_hybrid_model(input_shape):
    inputs = Input(shape=input_shape)

    # --- BLOCK 1: Feature Extraction (CNN) ---
    # "The Eye": Scans for micro-patterns (Wicks, Rejections)
    # filters=64: Looks for 64 different types of price shapes
    x = Conv1D(filters=64, kernel_size=3, padding='same', activation='swish')(inputs)
    x = BatchNormalization()(x)
    x = Dropout(0.2)(x)

    # --- BLOCK 2: Temporal Sequencing (Bi-LSTM) ---
    # "The Brain": Connects the shapes into a timeline
    x = Bidirectional(LSTM(64, return_sequences=True))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    # --- BLOCK 3: Context Awareness (Self-Attention) ---
    # "The Focus": Decides which part of the timeline matters MOST
    # This prevents the "memory overload" that caused your loss drift
    context = MultiHeadAttention(num_heads=4, key_dim=64)(x, x)
    x = Add()([x, context]) # Residual connection (Skip Connection)
    x = LayerNormalization()(x)
    
    # Global Average Pooling avoids the massive parameter explosion of Flatten()
    shared_features = GlobalAveragePooling1D()(x)
    
    # --- HEAD 1: Classification (Buy/Sell/Hold) ---
    # Swish helps gradient flow better than LeakyReLU in deep nets
    label_head = Dense(64, activation='swish')(shared_features)
    label_head = Dropout(0.3)(label_head)
    label_out = Dense(3, activation='softmax', name='label_output')(label_head)

    # --- HEAD 2: Regression (Next Return) ---
    return_head = Dense(32, activation='swish')(shared_features)
    return_out = Dense(1, activation='linear', name='return_output')(return_head)

    model = Model(inputs=inputs, outputs=[label_out, return_out])

    # --- SOTA COMPILATION ---
    # AdamW is mathematically superior for regularization than Adam + L2
    optimizer = AdamW(learning_rate=0.0005, weight_decay=1e-4)
    
    model.compile(
        optimizer=optimizer,
        loss={
            'label_output': 'sparse_categorical_crossentropy', 
            # Huber Loss is less sensitive to outliers than MSE
            'return_output': 'huber' 
        },
        loss_weights={'label_output': 1.0, 'return_output': 0.5},
        metrics={'label_output': 'accuracy', 'return_output': 'mae'}
    )
    return model

# --- 2. INITIALIZE ---
model = build_sota_hybrid_model((X_train.shape[1], X_train.shape[2]))
print("ğŸš€ SOTA Hybrid Model Built.")
model.summary()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# 1. Load Data
print("ğŸ•µï¸ Starting Feature Audit...")
df = pd.read_parquet("../data/EURUSD_H1_SOTA_Features.parquet")

# 2. Define Target (Next Hour Return)
# We check correlation against the RAW return, not the class label
target = df['target_next_return']
features = df.drop(columns=['label', 'target_next_return'])

# 3. Calculate Information Coefficient (IC)
# Spearman Correlation assesses monotonic relationships (ranking)
ic_scores = []
print(f"ğŸ“Š Scanning {len(features.columns)} features...")

for col in features.columns:
    # Drop NaNs to avoid errors
    valid_idx = features[col].notna() & target.notna()
    if valid_idx.sum() < 100: continue
    
    corr, p_val = spearmanr(features.loc[valid_idx, col], target.loc[valid_idx])
    ic_scores.append({'Feature': col, 'IC': corr, 'P-Value': p_val})

ic_df = pd.DataFrame(ic_scores)
ic_df['Abs_IC'] = ic_df['IC'].abs()
ic_df = ic_df.sort_values('Abs_IC', ascending=False)

# 4. The Verdict Visualization
plt.figure(figsize=(12, 6))
sns.barplot(x='IC', y='Feature', data=ic_df.head(20), palette='coolwarm')
plt.title("Top 20 Features by Predictive Power (Information Coefficient)")
plt.xlabel("Spearman Correlation with Next Hour Return")
plt.axvline(x=0.02, color='green', linestyle='--', label='Min Viable Signal (0.02)')
plt.axvline(x=-0.02, color='green', linestyle='--')
plt.legend()
plt.show()

# 5. Text Report
print("\nğŸ† --- TOP 10 PREDICTORS ---")
print(ic_df.head(10)[['Feature', 'IC', 'P-Value']])

print("\nğŸ’€ --- THE REALITY CHECK ---")
mean_ic = ic_df['Abs_IC'].mean()
best_ic = ic_df['Abs_IC'].max()
print(f"Average Feature Strength: {mean_ic:.4f}")
print(f"Best Feature Strength:    {best_ic:.4f}")

if best_ic < 0.02:
    print("âŒ CRITICAL FAIL: Your best feature has < 0.02 correlation.")
    print("   Action: You MUST change the target (e.g., predict H4 or Daily) or find new data.")
else:
    print("âœ… HOPE: You have some signal. We need to feature-select aggressively.")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
df = pd.read_parquet("../data/EURUSD_H1_SOTA_Features.parquet")

# 2. "The Lobotomy" - SELECT ONLY TOP 10 FEATURES
# Based on your Feature Audit image
top_features = [
    'log_ret',      # Momentum / Mean Reversion
    'dist_sma20',   # Trend Deviation
    'is_asia',      # Market Regime
    'is_dead_zone', # Volatility Regime
    'hour_cos',     # Time Cyclic
    'rsi_z',        # Overbought/Oversold
    'dist_sma50',   # Longer Trend
    'is_overlap',   # High Volatility Regime
    'is_london',    # Session
    'is_ny'         # Session
]

print(f"ğŸ“‰ Reducing Features from {df.shape[1]} to {len(top_features)}...")
X = df[top_features]
y = df['label'] + 1 # Encode -1,0,1 to 0,1,2

# 3. Split (Maintain Time Order!)
split_idx = int(len(X) * 0.8)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# 4. Scale (Crucial for Logistic Regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 5. Train Simple Linear Model
# class_weight='balanced' forces the model to care about Buy/Sell even if Hold is dominant
model = LogisticRegression(class_weight='balanced', C=0.01, solver='lbfgs', multi_class='multinomial')

print("ğŸ§  Training Logistic Regression...")
model.fit(X_train_scaled, y_train)

# 6. Evaluate
y_pred = model.predict(X_test_scaled)

print("\nğŸ“Š --- LOGISTIC REGRESSION PERFORMANCE ---")
print(classification_report(y_test, y_pred, target_names=['Sell', 'Hold', 'Buy']))

# 7. Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues',
            xticklabels=['Sell', 'Hold', 'Buy'], yticklabels=['Sell', 'Hold', 'Buy'])
plt.title("Logistic Regression (Top 10 Features Only)")
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

# 8. Inspect Coefficients (The "Why")
coefs = pd.DataFrame({
    'Feature': top_features,
    'Buy_Weight': model.coef_[2],  # Coefficients for "Buy" class
    'Sell_Weight': model.coef_[0]  # Coefficients for "Sell" class
})
print("\nâš–ï¸ --- MODEL LOGIC (Coefficients) ---")
print(coefs.sort_values('Buy_Weight', ascending=False))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Load Data
df = pd.read_parquet("../data/EURUSD_H1_SOTA_Features.parquet")

# 2. Define The "Raw Alpha" Strategy (Mean Reversion)
# Logic: If the market moved FAST in one direction, bet on it snapping back.

# Calculate thresholds (Top/Bottom 10% of movements)
upper_threshold = df['log_ret'].quantile(0.90)
lower_threshold = df['log_ret'].quantile(0.10)

print(f"ğŸ“‰ Short Threshold (Log Ret >): {upper_threshold:.5f}")
print(f"ğŸ“ˆ Long Threshold (Log Ret <):  {lower_threshold:.5f}")

# Generate Signals (1 = Buy, -1 = Sell, 0 = Hold)
# We "Fade" the move: High Return -> Sell (-1), Low Return -> Buy (1)
df['signal'] = 0
df.loc[df['log_ret'] > upper_threshold, 'signal'] = -1
df.loc[df['log_ret'] < lower_threshold, 'signal'] = 1

# 3. Calculate Strategy Returns
# Strategy Return = Signal * Next Hour Return
# (Shifted because 'target_next_return' matches the current row's action)
df['strategy_return'] = df['signal'] * df['target_next_return']

# Account for "Spread/Cost" (Rough approximation: 1 pip per trade = 0.0001)
cost_per_trade = 0.0001
df.loc[df['signal'] != 0, 'strategy_return'] -= cost_per_trade

# 4. Compare vs "Buy & Hold"
df['cumulative_market'] = (1 + df['target_next_return']).cumprod()
df['cumulative_strategy'] = (1 + df['strategy_return']).cumprod()

# 5. Visualize " The Truth"
plt.figure(figsize=(12, 6))
plt.plot(df['cumulative_market'], label='Buy & Hold EUR/USD', color='gray', alpha=0.5)
plt.plot(df['cumulative_strategy'], label='Mean Reversion Alpha (Top 10% Fades)', color='purple', linewidth=2)
plt.title("Alpha Verification: Raw Mean Reversion Strategy (H1)")
plt.xlabel("Hours")
plt.ylabel("Cumulative Return (1.0 = Breakeven)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 6. Stats
total_trades = df[df['signal'] != 0].shape[0]
final_return = df['cumulative_strategy'].iloc[-1]
win_rate = (df[df['signal'] != 0]['strategy_return'] > 0).mean()

print(f"ğŸ“Š Total Trades: {total_trades}")
print(f"ğŸ† Final Multiple: {final_return:.4f}x")
print(f"ğŸ¯ Win Rate: {win_rate:.2%}")

if final_return > 1.05:
    print("âœ… VERDICT: The Alpha is REAL. We just need to simplify the ML to follow this logic.")
else:
    print("âŒ VERDICT: The Alpha is an ILLUSION. H1 Noise is killing the edge. Move to H4.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# 1. Load H1 Data & Resample to H4 (Standard Procedure)
df_h1 = pd.read_parquet("../data/EURUSD_H1_SOTA_Features.parquet")
df = df_h1.iloc[::4].copy()
df['target_next_return'] = df['log_ret'].shift(-1)
df = df.dropna()

# 2. Define Target: Simple Direction (1 = Up, 0 = Down)
# We ditch the "Hold" class. Let's just predict Up/Down probability.
df['target_class'] = (df['target_next_return'] > 0).astype(int)

# 3. Select Features (The ones with high IC from your audit)
features = [
    'dist_sma50', 'dist_sma20', 'hour_cos', 'is_dead_zone', 
    'rsi_z', 'is_asia', 'is_london', 'log_ret', 
    'is_ny', 'bb_width', 'hour_sin', 'vol_qk'
]

X = df[features]
y = df['target_class']

# 4. Walk-Forward Split (Train on Past, Test on Future)
# We use the last 20% for testing (approx 2 years of H4 bars)
split = int(len(X) * 0.8)
X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]

# 5. Train Random Forest
# n_estimators=100: Build 100 trees
# min_samples_leaf=20: Prevent overfitting (don't memorize single candles)
# class_weight='balanced': Handle any Up/Down imbalance
model = RandomForestClassifier(
    n_estimators=100, 
    max_depth=7, 
    min_samples_leaf=20, 
    random_state=42, 
    n_jobs=-1,
    class_weight='balanced_subsample'
)

print("ğŸŒ² Training H4 Random Forest...")
model.fit(X_train, y_train)

# 6. Evaluate
print("ğŸ§  Generating Predictions...")
# Get probability of "Up" (Class 1)
probs = model.predict_proba(X_test)[:, 1] 

# 7. The "Confident" Strategy
# Only trade if the model is > 55% sure or < 45% sure
threshold = 0.55
signals = np.zeros(len(probs))
signals[probs > threshold] = 1   # Buy
signals[probs < (1 - threshold)] = -1  # Sell

# 8. Backtest the RF Strategy
# returns = signal * actual_return
# We align signals with the test set's future returns
market_returns = df['target_next_return'].iloc[split:]
strategy_returns = signals * market_returns

# Deduct Spread (0.0001 per trade)
strategy_returns[signals != 0] -= 0.0001

# Cumulative Curves
cum_market = (1 + market_returns).cumprod()
cum_strategy = (1 + strategy_returns).cumprod()

# 9. Visualize
plt.figure(figsize=(12, 6))
plt.plot(cum_market.values, label='Buy & Hold', color='gray', alpha=0.5)
plt.plot(cum_strategy.values, label='Random Forest (H4 Conditional)', color='green', linewidth=2)
plt.title(f"H4 Random Forest Backtest (Threshold: {threshold})")
plt.xlabel("H4 Bars (Test Set)")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 10. Stats
trades = np.sum(signals != 0)
win_rate = np.mean(strategy_returns[signals != 0] > 0)
final_equity = cum_strategy.iloc[-1]

print(f"ğŸ“Š Total Trades: {trades}")
print(f"ğŸ¯ Win Rate: {win_rate:.2%}")
print(f"ğŸ’° Final Equity: {final_equity:.4f}x")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# 1. Load H1 Data
df_h1 = pd.read_parquet("../data/EURUSD_H1_SOTA_Features.parquet")

# 2. Resample to H4 (Base timeframe)
df = df_h1.iloc[::4].copy()

# 3. Create Multiple Targets (Future Returns)
# We calculate returns for different horizons ahead
horizons = [1, 2, 3, 6, 12] # in H4 bars -> 4h, 8h, 12h, 24h, 48h
horizon_labels = ['4H', '8H', '12H', '24H', '48H']

for h in horizons:
    # Log return over next 'h' bars
    # Summing log returns = Log(Total Return)
    df[f'ret_{h*4}h'] = df['log_ret'].rolling(window=h).sum().shift(-h)

df = df.dropna()

# 4. The "Horizon Audit"
# We check which horizon has the strongest signal for 'dist_sma50'
feature = 'dist_sma50'
corrs = []

for h, label in zip(horizons, horizon_labels):
    target_col = f'ret_{h*4}h'
    corr, _ = spearmanr(df[feature], df[target_col])
    corrs.append(corr)

# 5. Visualize
plt.figure(figsize=(10, 5))
# We plot Absolute Correlation because we care about magnitude, not sign
sns.barplot(x=horizon_labels, y=np.abs(corrs), palette='magma')
plt.title(f"Signal Strength vs. Time Horizon (Feature: {feature})")
plt.ylabel("Spearman Correlation (Absolute)")
plt.xlabel("Prediction Horizon")
plt.axhline(y=0.03, color='green', linestyle='--', label='Tradable Threshold')
plt.legend()
plt.show()

# 6. Report
best_idx = np.argmax(np.abs(corrs))
best_horizon = horizon_labels[best_idx]
best_score = np.abs(corrs)[best_idx]

print(f"ğŸ† Best Time Horizon: {best_horizon}")
print(f"ğŸ’ª Signal Strength:   {best_score:.4f}")

if best_score > 0.05:
    print("âœ… VERDICT: Pivot to Swing Trading. Predict the next 24h-48h.")
else:
    print("âŒ VERDICT: The signal is weak across all horizons. Data Failure.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# 1. Load H1 Data & Resample to H4
print("ğŸ”„ Preparing '12-Hour Sniper' Data...")
df_h1 = pd.read_parquet("../data/EURUSD_H1_SOTA_Features.parquet")
df = df_h1.iloc[::4].copy() # H4 Resampling

# 2. ENGINEER THE "SWEET SPOT" TARGET (12-Hour Return)
# We sum the log returns of the next 3 bars (3 * 4h = 12h)
# rolling(3).sum() sums the *trailing* 3, so we shift back by 3 to get *future* 3
df['target_12h_return'] = df['log_ret'].rolling(window=3).sum().shift(-3)
df = df.dropna()

# 3. Define Class: 1 = Up > 10 pips, -1 = Down < -10 pips, 0 = Chop
# We add a "Buffer" so the model doesn't sweat the small stuff
buffer = 0.0010 # 10 pips
df['target_class'] = 0
df.loc[df['target_12h_return'] > buffer, 'target_class'] = 1
df.loc[df['target_12h_return'] < -buffer, 'target_class'] = -1

# 4. Features (The Rubber Band + Timing)
features = [
    'dist_sma50', 'dist_sma20', 'rsi_z', # The "Stretch"
    'hour_cos', 'is_london', 'is_ny',    # The "Timing" (Liquidity)
    'vol_qk'                             # The "Fuel"
]

X = df[features]
y = df['target_class']

# 5. Train/Test Split (Last 2 years)
split = int(len(X) * 0.8)
X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]

# 6. Train Random Forest (Optimized for Multi-Class)
model = RandomForestClassifier(
    n_estimators=200, 
    max_depth=6,         # Keep it shallow to prevent overfitting
    min_samples_leaf=50, # Force it to find broad patterns
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

print(f"ğŸŒ² Training Sniper on {len(X_train)} samples...")
model.fit(X_train, y_train)

# 7. Generate Predictions (Probabilities)
probs = model.predict_proba(X_test)

# 8. Strategy Logic (High Confidence Only)
# We only trade if model is > 50% sure of a Direction (1 or -1)
# Note: probs is [prob_-1, prob_0, prob_1]
threshold = 0.45 # Lower threshold because 3-class probability is harder

signals = np.zeros(len(X_test))
signals[probs[:, 2] > threshold] = 1   # High confidence Buy
signals[probs[:, 0] > threshold] = -1  # High confidence Sell

# 9. Backtest
# Returns = Signal * Actual 12H Return
market_returns = df['target_12h_return'].iloc[split:]
strategy_returns = signals * market_returns

# Deduct Spread (0.0001 per trade)
strategy_returns[signals != 0] -= 0.0001

# 10. Visualize
cum_market = (1 + market_returns).cumprod()
cum_strategy = (1 + strategy_returns).cumprod()

plt.figure(figsize=(12, 6))
plt.plot(cum_market.values, label='Market (12H Returns)', color='gray', alpha=0.3)
plt.plot(cum_strategy.values, label='12H Sniper Strategy', color='gold', linewidth=2)
plt.title(f"12-Hour Sniper Backtest (Target: 3-Bar Cumulative)")
plt.xlabel("Trade Opportunities (H4 Bars)")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 11. Stats
trades = np.sum(signals != 0)
win_rate = np.mean(strategy_returns[signals != 0] > 0)
final_eq = cum_strategy.iloc[-1]

print(f"ğŸ“Š Trades: {trades}")
print(f"ğŸ¯ Win Rate: {win_rate:.2%}")
print(f"ğŸ’° Final Equity: {final_eq:.4f}x")

if final_eq > 1.05:
    print("âœ… VERDICT: WE FOUND IT. Deploy this model.")
else:
    print("âŒ VERDICT: Even the 12H signal is untradable. Project relies on Alternative Data.")
```
